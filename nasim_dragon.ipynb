```python
channel_volume = pd.read_csv("./Data/Channel_Volume.csv")
channel_volume.head(10)
```

```python
channel_volume.replace(to_replace="000 litres",value="million litres")
channel_volume['Year_date'] = pd.to_datetime(channel_volume['Year_date'])

channel_volume['Year_date'] = channel_volume['Year_date'].dt.strftime('%d-%m-%y')


```

```python
channel_volume['Volume'] = channel_volume['Volume'].apply(lambda x: float(x.replace(',', '')))

channel_volume.head(10)
```

```python
nan_count = channel_volume.isna().sum().sum()
print("Number of NaN values:", nan_count)
```
